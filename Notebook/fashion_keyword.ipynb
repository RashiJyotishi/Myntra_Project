```python
import re
from collections import Counter
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')

# Sample caption data
captions = [
    '''Thumbelina bow earrings at Ã¢â€šÂ¹450/-
 The product is Anti- tarnish, water resistant, hypoallergenic and stainless steel.

 #pinterest #bowearrings #trendyfashion #genzfashion''',
    '''We are open to take your order!!!!!

Itâ€™s time to order T-shirts with your favorite inscriptions ðŸ¥°.

Kindly send us a DM to place your order!

Our T-shirts are available for as low as 8k
S | M | L | XL | XXL
ãƒ»
ãƒ»
ãƒ»
ãƒ»
ãƒ»
#1123collections #shirtsvendor #shirtfreaksquad #shirtfreak #myshirtfreakstyle #customshirts #genzfashion #genz #naijafashiongallery #naijafashiontrends #shirtsforsale #shirtshop #customizedShirt #customizedshirts #customizedshirtsph #customizedchristiantshirts #fashiongoalsz #faithbasedbusiness #faithbasedapparel #faithbasedclothing #faithbased''',
    "Vintage vibes with this outfit. #vintage #style",
    # Add more captions
]

# Preprocess captions
def preprocess_caption(caption):
    caption = re.sub(r'http\S+', '', caption)  # Remove URLs
    caption = re.sub(r'[^A-Za-z0-9\s#]', '', caption)  # Remove special characters
    caption = caption.lower()  # Convert to lowercase
    return caption

# Tokenize and remove stop words
def tokenize_caption(caption):
    stop_words = set(stopwords.words('english'))
    tokens = nltk.word_tokenize(caption)
    tokens = [token for token in tokens if token not in stop_words and token != '']
    return tokens

# Extract keywords
all_tokens = []
for caption in captions:
    processed_caption = preprocess_caption(caption)
    tokens = tokenize_caption(processed_caption)
    all_tokens.extend(tokens)

# Count word frequencies
word_freq = Counter(all_tokens)

# Display top keywords
top_keywords = word_freq.most_common(10)
print("Top Keywords:", top_keywords)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Preprocess captions
processed_captions = [preprocess_caption(caption) for caption in captions]

# Apply TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(processed_captions)
feature_names = vectorizer.get_feature_names_out()

# Get top keywords for each caption
def get_top_tfidf_words(row, features, top_n=5):
    top_indices = row.argsort()[-top_n:][::-1]
    top_words = [features[i] for i in top_indices]
    return top_words

top_keywords_per_caption = []
for i in range(tfidf_matrix.shape[0]):
    row = tfidf_matrix.getrow(i).toarray()[0]
    top_keywords = get_top_tfidf_words(row, feature_names)
    top_keywords_per_caption.append(top_keywords)

# Display top keywords for each caption
for i, keywords in enumerate(top_keywords_per_caption):
    print(f"Caption {i+1} Top Keywords:", keywords)
```

```python
import requests
from PIL import Image
import numpy as np
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import re
import nltk
from nltk.corpus import stopwords
from io import BytesIO
import cairosvg

# Download required NLTK data
nltk.download('stopwords')

# Sample captions data
captions = [
  "Coquette Lace Pointelle Tees",
  "Bow-Detail Jumpers",
  "Mary Jane Flats",
  "Statement Scarves",
  "Mini Cardigans",
  "Vintage Cardigans",
  "Pearls",
  "Mismatched Patterns",
  "Eco-friendly Linen Clothes",
  "Robotic-looking Fabrics",
  "Ruffled Sleeves",
  "Bell Sleeves",
  "Puff Sleeves",
  "Pink Lacy Details",
  "Bows",
  "Pastels",
  "Floral Prints",
  "Long Dresses",
  "Oversized Jackets",
  "Wide-Leg Pants",
  "Cropped Blazers",
  "Patchwork Jeans",
  "Cargo Pants",
  "Sheer Tops",
  "Holographic Fabrics",
  "Faux Fur Coats",
  "Pleated Skirts",
  "Graphic Tees",
  "Oversized Hoodies",
  "Denim Overalls",
  "Fringe Jackets",
  "Chunky Sneakers",
  "Bucket Hats",
  "Platform Boots",
  "Velvet Dresses",
  "Utility Vests",
  "Flared Pants",
  "Animal Print Accessories",
  "Knitted Vests",
  "Transparent Tote Bags"
]

# Preprocess captions
def preprocess_caption(caption):
    caption = re.sub(r'http\S+', '', caption)  # Remove URLs
    caption = re.sub(r'[^A-Za-z0-9\s#]', '', caption)  # Remove special characters
    caption = caption.lower()  # Convert to lowercase
    return caption

# Tokenize and remove stop words
def tokenize_caption(caption):
    stop_words = set(stopwords.words('english'))
    tokens = nltk.word_tokenize(caption)
    tokens = [token for token in tokens if token not in stop_words and token != '']
    return tokens

# Combine all captions into one text
all_tokens = []
for caption in captions:
    processed_caption = preprocess_caption(caption)
    tokens = tokenize_caption(processed_caption)
    all_tokens.extend(tokens)

# Join tokens to create a single text
text = ' '.join(all_tokens)

# Download the SVG image from the link
svg_url = "https://cdn.worldvectorlogo.com/logos/myntra-1.svg"
response = requests.get(svg_url)

# Convert SVG to PNG using cairosvg
png_data = cairosvg.svg2png(bytestring=response.content)

# Load the PNG image using PIL
image = Image.open(BytesIO(png_data))

# Convert the image to a NumPy array
mask = np.array(image)
def myntra_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    # Define Myntra's primary colors (orange and pink)
    colors = ['#FF3E6C', '#FFAB00']  # Pink and Orange

    # Return a color based on word index to ensure different colors for different words
    return colors[hash(word) % len(colors)]

# Generate word cloud with the custom mask and color function
wordcloud = WordCloud(width=800, height=400, background_color='white', mask=mask,
                      color_func=myntra_color_func, contour_color='black', contour_width=2).generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
