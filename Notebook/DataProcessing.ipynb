```python
from googletrans import Translator
```

```python
import time
translator = Translator()

# Replace 'caption_column' with the name of your column
caption_column = 'Caption'

# Function to translate text with error handling
def translate_text(text):
    try:
        translated = translator.translate(text, dest='en').text
        return translated
    except Exception as e:
        print(f"Error translating text: {e}")
        return text

# Translate each caption in the column with a delay
for index, row in df.iterrows():
    if pd.notna(row[caption_column]):
        df.at[index, caption_column] = translate_text(str(row[caption_column]))
        time.sleep(1)  # Adding delay to avoid hitting the rate limit

# Display the first few rows of the DataFrame
print(df.head())
```

# Keywords Extraction

1. To analyses Captions, we tokenized them, removed unwanted characters and then removed the stop words to get only keywords.
2. The next in the plan would be the categorise thosa words and labeling them.

```python
def clean_text_and_remove_stopwords(text):
    # Remove unwanted characters and convert to lowercase
    cleaned_text = re.sub(r'[^A-Za-z\s]', '', str(text).lower())
    # Tokenize the cleaned text into words
    words = word_tokenize(cleaned_text)
    # Remove numbers
    words = [word for word in words if not word.isdigit()]
    # Remove stop-words
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word not in stop_words]
    # Join the remaining words into a single string
    return ' '.join(filtered_words)

# Apply the cleaning and stop-word removal function to the caption column
df[caption_column] = df[caption_column].apply(clean_text_and_remove_stopwords)

# Display the first few rows of the cleaned DataFrame
print(df.head())
```

```python
def translate_comments(comments):
    translator = Translator()
    translated_comments = []
    for comment in comments:
        if isinstance(comment, str):  # Check if comment is a string (handle non-array cases)
            translated_text = translator.translate(comment, dest='en').text
            translated_comments.append(translated_text)
        else:
            translated_comments.append(None)  # Handle non-string cases (if any)
    return translated_comments

# Example preprocessing: Remove unwanted characters like \n, \r, quotes, etc.
df2['Comment desc'] = df2['Comment desc'].str.replace(r'\r\n', ' ').str.replace(r'["\[\]]', '')

# Apply translation function to 'Comment desc' column
df2['Translated_Comments'] = df2['Comment desc'].apply(lambda x: translate_comments(eval(x)) if pd.notna(x) else [])

# Display the DataFrame to verify
print(df2.head())
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
import numpy as np
nltk.download('vader_lexicon')
analyzer = SentimentIntensityAnalyzer()

# Function to apply sentiment analysis using VADER
def analyze_sentiment(comments):
    if isinstance(comments, list):
        scores = [analyzer.polarity_scores(comment)['compound'] for comment in comments]
        return scores
    else:
        return []
# Apply sentiment analysis to 'Translated_Comments' column
combined_df['Sentiment_Scores'] = combined_df['Translated_Comments'].apply(analyze_sentiment)

# Calculate mean sentiment score for each row
combined_df['Sentiment_Score'] = combined_df['Sentiment_Scores'].apply(lambda scores: np.mean(scores))
```
